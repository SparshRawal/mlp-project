```python
def visualize(original, augmented):
    # fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Original image')
    plt.imshow(original)

    plt.subplot(1,2,2)
    plt.title('Augmented image')
    
    plt.imshow(augmented)
    flipped = tf.image.flip_left_right(original)
    visualize(original, flipped)
```
